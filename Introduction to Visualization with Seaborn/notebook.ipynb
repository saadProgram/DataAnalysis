```python
# Importing course packages; you can add more too!
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Importing course datasets as DataFrames
country_data = pd.read_csv('datasets/countries-of-the-world.csv', decimal=",")
mpg = pd.read_csv('datasets/mpg.csv')
student_data = pd.read_csv('datasets/student-alcohol-consumption.csv', index_col=0)
survey = pd.read_csv('datasets/young-people-survey-responses.csv', index_col=0)

sns.set_context('talk')

survey.head() # Display the first five rows of this DataFrame
```

```python
sns.set_style('darkgrid')
g = sns.relplot(x='GDP ($ per capita)' , y='Literacy (%)', data=country_data,  kind='scatter', alpha=0.6, hue='Region')
g.fig.suptitle("Relation of GDP per Capita vs. Literacy in each Region", y=1.04)
plt.show()
```

```python
plt.clf()
f = sns.relplot(y='weight', x='model_year', data=mpg, kind='line', hue='origin', style='origin', markers=True, dashes=False)
f.fig.suptitle("Weight of the cars in each Origin over the Model Years", y=1.04)
f.set(xlabel='Model Year', ylabel='Weight')
plt.show()
```

```python
plt.clf()
col_order = ['<2 hours', '2 to 5 hours', '5 to 10 hours', '>10 hours']
s = sns.catplot(x='failures', y='G3', kind='box', data=student_data, col='study_time' , col_wrap=2, col_order=col_order)
s.fig.suptitle('The relationship of no. of failures with Final Grades, with respect to study time', y=1.03)
s.set_titles('Study Time: {col_name}')
s.set(xlabel='No. of Failures', ylabel='Final Grades')
plt.show()
```

```python
plt.clf()
sns.set_style('white')
fig = sns.catplot(y='Loneliness', x='Internet usage', data=survey, kind='bar', col='Gender', palette="Blues")
fig.set_xticklabels(rotation=35, fontsize=12)
fig.fig.suptitle('Internet Usage with Loneliness of People by Gender', y=1.03)
fig.set_titles('Gender is {col_name}')
plt.show()
```
